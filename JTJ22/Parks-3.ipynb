```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

parks_data = pd.read_csv("fa329e3d-89ff-4708-8ddf-81bfedcad11d.csv", index_col="id" ,engine = 'python')
```

```python
fig = plt.figure()
fig.patch.set_facecolor('black')
fig.patch.set_alpha(0.2)

npn = parks_data['neighborhood'].value_counts().plot(kind='barh', figsize=(25,25), title='Number of Parks Per Neighborhood', fontsize=12,
                                             color='g', edgecolor='black', linewidth=2)
npn.set(facecolor = 'tan')
npn.patch.set_alpha(0.4)
```

```python
park1 = pd.read_csv("fa329e3d-89ff-4708-8ddf-81bfedcad11d.csv", index_col="id")
park2 = pd.read_csv("https://data.wprdc.org/datastore/dump/bb57d0a7-e8ee-4218-8906-0dedc903038c", index_col="objectid")
```

```python
park = pd.merge(left=park1, right=park2, left_on='name', right_on='updatepknm')
park = park.filter(['neighborhood', 'name', 'acreage'])
parktotal = park.groupby('neighborhood').sum()
parktotal
```

```python
npn = parktotal.sort_values('acreage').plot(kind='barh', figsize=(25,25), title='Acreage of Parks Per Neighborhood', fontsize=12,
                                             color='g', edgecolor='black', linewidth=2)
npn.set(facecolor = 'tan')
npn.patch.set_alpha(0.4)
```
